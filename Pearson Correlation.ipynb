Import dependencies

```python
import os
import pandas as pd
```

Read the filtered dataset named 'combined_light.csv' and create a pandas dataframe.

```python
print('Creating the dataframe from combined_light.csv file..')
df = pd.read_csv('./data_input/combined_light.csv', sep=',')

df.date = pd.to_datetime(df.date)
print('Done.\n')

df
```

```python
df_title = pd.read_csv('./data_input/movie_titles.csv', sep=',', names=['movie_id','year','title'])
df_title
```

```python
result = df.merge(df_title, how = "inner")
result
```

```python
df = result.pivot_table(index='customer_id', columns='title', values='rating')
df
```

```python
x = df['2 Fast 2 Furious']
corr = df.corrwith(x)
corr.sort_values(ascending = False).head(10)
```
